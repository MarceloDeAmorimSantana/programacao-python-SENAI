```python
# Código 01 - Escreva um programa que solicite ao usuário um número inteiro e verifique
# se ele é um número positivo. Em seguida, exiba a mensagem correspondente.

numInt = int(input("\nDigite o 1º número inteiro -> "))

if numInt < 0:
    print(f'\n[ {numInt} ] é um número NEGATIVO!')

elif numInt > 0:
    print(f'\n[ {numInt} ] é um número POSITIVO!')

else:
    print(f'\n[ {numInt} ] é um número NEUTRO!')
```

```python
# Código 04 - Escreva um programa que solicite ao usuário um número inteiro e verifique se ele é um número perfeito.
# Um número perfeito é aquele cuja soma dos seus divisores (excluindo ele mesmo) é igual ao próprio número.
# Por exemplo, 6 é um número perfeito, pois seus divisores são 1, 2 e 3, e 1 + 2 + 3 = 6.

numInt = int(input("Digite um número pra verificar se ele é um Nº perfeito -> "))

QtdSomaDivisoes = 1

if ( numInt % 2 ) == 0:
    QtdSomaDivisoes = QtdSomaDivisoes + 2
    
if ( numInt % 3 ) == 0:
    QtdSomaDivisoes = QtdSomaDivisoes + 3


if QtdSomaDivisoes == numInt:
    print('\nO número',numInt,'é um número perfeito!')
    
else:
    print('\nO número',numInt,'não é um número perfeito!')
    
    
```

```python
# Código 07 -  Escreva um programa que solicite ao usuário um número inteiro e verifique se ele é um quadrado perfeito.
# Um número é considerado um quadrado perfeito quando sua raiz quadrada é um número inteiro.

num = int(input('\nDigite um número inteiro -> '))

raiz = num ** 0.5

if raiz % 1 != 0:
    print(f'\nO número {num}, não é um quadrado perfeito, pois sua raiz quadrada é [',raiz,'] que não é um Nº inteiro!')
else:
    print(f'\nO número {num}, é um quadrado perfeito, pois sua raiz quadrada é [',raiz,'] que é um Nº inteiro!')
    
```

```python
# Código 09 - Faça um Programa que leia 2 números e em seguida 
# pergunte ao usuário qual operação ele deseja realizar.
# O resultado da operação deve ser acompanhado de uma frase que diga se o número é:

#----------leia 2 números----------#

num1 = float(input("\nDigite o 1º número -> "))
num2 = float(input("\nDigite o 2º número -> "))
resultado = 0.0

operacao = input("\n-----------------------------Escolha uma Operação:-----------------------------                 \n[A]-Adição | [S]-Subtração | [M]-Multiplicação | [D]-Divisão | [P]-Potenciação.\n\n")
operacao = operacao.lower()


#----------Perguntar ao usuário qual operação ele deseja realizar----------#

if operacao == "a":
    resultado = num1 + num2
    print('\nO resultado da soma é = ',resultado)
    
elif operacao == "s":
    resultado = num1 - num2
    print('\nO resultado da subtração é = ',resultado)

elif operacao == "m":
    resultado = num1 * num2
    print('\nO resultado da multiplicação é = ',resultado)    

elif operacao == "d":
    resultado = num1 / num2
    print('\nO resultado da divisão é = ',resultado)    
                 
elif operacao == "p":
    resultado = num1 ** num2
    print('\nO resultado da exponenciação é = ',resultado)    

else:
    print('\nVocê digitou [',operacao,']!! Favor selecione uma operação válida!')


#----------Analisar se é par ou ímpar----------#

if resultado % 2 == 0:
    print('\n[',resultado,'] é um número PAR!')
    
else:
    print('\n[',resultado,'] é um número IMPAR!')


#----------Analisar se é positivo ou negativo----------#

if resultado < 0:
    print('\n[',resultado,'] é um número NEGATIVO!')

elif resultado > 0:
    print('\n[',resultado,'] é um número POSITIVO!')

else:
    print('\n[',resultado,'] é um número NEUTRO')
    

    
#----------Inteiro ou decimal----------# 

if resultado % 1 != 0:
     print('\nO número [',resultado,'] é DECIMAL')   
        
else:
     print('\nO número [',resultado,'] é INTEIRO')



# numero = resultado
# verifica = isinstance(numero,float)
# if verifica == True:
#     print('\nO número [',resultado,'] é DECIMAL')
# else:
#     print('\nO número [',resultado,'] é INTEIRO')
# O código acima sempre exibirá DECIMAL (devido a conversão inicial), por isso não foi utilizado!
    
```
